```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June']
developers_2019 = [1417446, 1414166, 1602974, 1601502, 1591965, 1479098]
developers_2022 = [2360303, 2312232, 2534641, 2539777, 2573442, 2456974]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(months, developers_2019, label='Developers 2019', marker='o', color='blue')
plt.plot(months, developers_2022, label='Developers 2022', marker='o', color='green')
plt.title('Number of Developers in First 6 Months on Github (2019 vs 2022)')
plt.xlabel('Month')
plt.ylabel('Number of Developers')
plt.legend()
plt.grid(True)

plt.show()
```

```python
total_pushes_2019 = 265993271
total_pushes_2022 = 554364774
```

```python
plt.figure(figsize=(8, 5))
bar_width = 0.35
index = np.arange(2)

plt.bar(index, [total_pushes_2019, total_pushes_2022], bar_width, color=['blue', 'green'], label='Total Pushes')

plt.xlabel('Year')
plt.ylabel('Total Pushes')
plt.title('Total Pushes in First 6 Months (2019 vs 2022)')
plt.xticks(index, ['2019', '2022'])
plt.legend()
plt.grid(True)

plt.show()
```

```python
pushes_per_developer_2019 = total_pushes_2019 / np.mean(developers_2019)
pushes_per_developer_2022 = total_pushes_2022 / np.mean(developers_2022)
```

```python
years = ['2019', '2022']
pushes_per_developer = [pushes_per_developer_2019, pushes_per_developer_2022]
```

```python
plt.figure(figsize=(8, 5))
bar_width = 0.35
index = np.arange(2)

plt.bar(index, pushes_per_developer, bar_width, color=['blue', 'green'], label='Pushes per Developer')

plt.xlabel('Year')
plt.ylabel('Pushes per Developer')
plt.title('Average Pushes per Developer (2019 vs 2022)')
plt.xticks(index, years)
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()

print("Pushes per developer in 2019:", pushes_per_developer_2019)
print("Pushes per developer in 2022:", pushes_per_developer_2022)
```

```python
issues_opened_2019 = 5464656
issues_closed_2019 = 3532295
issues_opened_2022 = 7459185
issues_closed_2022 = 4865575
```

```python
opened_issues = [issues_opened_2019, issues_opened_2022]
closed_issues = [issues_closed_2019, issues_closed_2022]
```

```python
bar_width = 0.35
index = np.arange(2)

plt.figure(figsize=(10, 6))

plt.bar(index, opened_issues, bar_width, color='blue', label='Issues Opened')

plt.bar(index + bar_width, closed_issues, bar_width, color='green', label='Issues Closed')

plt.xlabel('Year')
plt.ylabel('Number of Issues')
plt.title('Issues Opened and Closed (2019 vs 2022)')
plt.xticks(index + bar_width / 2, years)
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
ratio_2019 = issues_closed_2019 / issues_opened_2019
ratio_2022 = issues_closed_2022 / issues_opened_2022
```

```python
ratio_2019 = round((issues_closed_2019 / issues_opened_2019) * 100, 3)
ratio_2022 = round((issues_closed_2022 / issues_opened_2022) * 100, 3)

years = ['2019', '2022']
ratios = [ratio_2019, ratio_2022]


plt.figure(figsize=(10, 6))


plt.barh(years, ratios, color=['blue', 'green'])

plt.xlabel('Ratio of Closed to Opened Issues (%)')
plt.title('Comparison of Closed to Opened Issues Ratio (2019 vs 2022)')
plt.xlim(0, 100)  # Setting x-axis limit to 100% for better visibility

for index, value in enumerate(ratios):
    plt.text(value + 1, index, f'{value:.3f}%', va='center')

plt.grid(True)
plt.tight_layout()
plt.show()


print(f"Ratio of Closed to Opened Issues in 2019: {ratio_2019}%")
print(f"Ratio of Closed to Opened Issues in 2022: {ratio_2022}%")
```
